```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols = df.select_dtypes(include='object').columns
for col in cols:
    df[col] = le.fit_transform(df[col])

print(df.head())
```

```python
df.info()
```

```python
# Job_ID column was previously dropped or handled, so removing redundant drop statement.

# Now that preprocessing is complete, define X and y, then perform the train-test split.
X = df.drop('Salary_LPA', axis=1)
y = df['Salary_LPA']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
lg_model = LinearRegression()
lg_model.fit(X_train, y_train)
y_pred_lg = lg_model.predict(X_test)
```

```python
# y_pred
```

```python
r2_score(y_test, y_pred_lg)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
r2 = r2_score(y_test, y_pred_lg)
mae = mean_absolute_error(y_test, y_pred_lg)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_lg))
print("R2 Score :", r2)
print("MAE      :", mae)
print("RMSE     :", rmse)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
rf_model = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)

print(r2_score(y_test, y_pred_rf))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
r2 = r2_score(y_test, y_pred_rf)
mae = mean_absolute_error(y_test, y_pred_rf)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_rf))
print("R2 Score :", r2)
print("MAE      :", mae)
print("RMSE     :", rmse)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)
print(r2_score(y_test, y_pred_dt))
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
r2 = r2_score(y_test, y_pred_dt)
mae = mean_absolute_error(y_test, y_pred_dt)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_dt))
print("R2 Score :", r2)
print("MAE      :", mae)
print("RMSE     :", rmse)
```
